# Example - How to generate calibration curves

Analyze HPLC calibration runs to obtain calibration curves for different substrates. All calibration curves are saved in a cetral json file: data/calibration.json.



```python
import sys, os
import glob 
import cremerlab.hplc #submodule to analyze hplc data

#load required pacakges (as always)
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.integrate as spi
from scipy.integrate import odeint #this is the module to solve ODEs
%matplotlib inline
# Load the scipy module
import scipy.stats 
import json
```

# Calibration glucose and butyrate detection
This shows an example where glucose and butyrate concentrations where both changes in different samples to get calibration curves for both substrates.

```python
#1. Specify name of folder where data is stored 
foldername='epsilon_glucose_butyrate_cal'

#2. Convert raw calibration files into simpler files which store chromatograms only
raw_calib_files = glob.glob('data/'+foldername+'/*.txt')
cremerlab.hplc.convert(raw_calib_files) #convert into simple csf files with chromatogram info (and some metadata as comments)
```

```python
#3. Plot chromatograms and autodetect positions (retention times) of peaks (to check chromatograms and see if autodetection and background subtraction works)
#set list of hplc chromatograms to analyze
cal_files = glob.glob('data/'+foldername+'/converted/*.csv') 

#set options background correction
backgroundsubstraction=True
backgroundsubstraction_iterations=80

#plot chromatograms of all calibration samples showing raw chromograms and the adjusted onces after background substraction. 
#autopeak detection included with peak positions shown ad dashed lines.
chroms, plot, cal_rettimes = cremerlab.hplc.batch_plot(cal_files,time_window=[14, 37], upper_limit=5000,lower_limit=-500, detection_peakpositions='peakautodetection',backgroundsubstraction=backgroundsubstraction,backgroundsubstraction_iterations=backgroundsubstraction_iterations)

#alternative: setting peak positions by hand
#peak_retentiontime=sorted(list(np.array([14,15.85,36])))
#and then plot chromatograms and these positions
#cal_chroms, cal_plot, cal_rettimes = cremerlab.hplc.batch_plot(cal_files,time_window=[14, 37], upper_limit=20000,lower_limit=-500, plot_verticallines=peak_retentiontime)
```

```python
#quantify areas of peaks (fitting of skewed Gaussians)

#start with converted files
cal_files = sorted(glob.glob('data/'+foldername+'/converted/*.csv'))
print(cal_files)

#set parameters for fitting skewed Gaussian to each peak
boundpars=[0.4,0.01,0.0,-0.1,1.5,0.01,.9,0.1]
#1. min. amplitude of peak height (multiplication of amplitude of detected peak position)
#2. min. position of peak (offset of detected peak position to the left)
#3. min. width of fitted peak
#4. min. value of skew parameter
#5. max. amplitude of peak height (multiplication of amplitude of detected peak position)
#6. max. position of peak (offset of detected peak position to the right)
#7. max. width of fitted peak
#8. min. value of skew parameter
#alternative: set boundpars=None to use standard settings

cal_chroms, cal_peaks, cal_plot = cremerlab.hplc.batch_process(cal_files, time_window=[14, 37], show_viz=True,upper_limit=50000,lower_limit=-500,manual_peak_positions=cal_rettimes,boundpars=boundpars,buffer=100,backgroundsubstraction=True,backgroundsubstraction_iterations=backgroundsubstraction_iterations)
```

```python
#print all peaks and the parameters of the fit
display(cal_peaks)
```

# Glucose calibration

To derive calibration curve, pick peak for a specic component and analyse how area of peak changes with concentration

```python
#start with giving the calibration a name which later can be used to analyze samples with this speciffic calibration
name='epsilon_glucose'
date='220331'
#specify which retention time range to use for calibration curve and select all peaks having this retention time
peak_retentiontime=[15., 16.]
calibration_peak = cal_peaks[(cal_peaks['retention_time'] >= peak_retentiontime[0]) & (cal_peaks['retention_time'] <= peak_retentiontime[1])]
display(calibration_peak)
```

```python
#read in concentrations for the different samples (this is based on name of sample)
concs = []

#loop through each sample 
for s in calibration_peak['sample'].values:
    # split the file name by underscores
    split = s.split('_')
    #print(split)
    ## Take only the last item that was split by an underscore and remove the mM
    concentration = split[1].split('mM')[0]
    ## Add the concentration to the storage list as a float
    print("detected concentration: "+concentration)
    concs.append(float(concentration))


#add the concentration as a new column and display results 
calibration_peak['concentration_mM'] = concs #warnin
display(calibration_peak)
```

```python
#plot calibration curve and fit a line

#select which samples to use
calibration_peak=calibration_peak.loc[calibration_peak['concentration_mM']<20]

#perform linear fit
output = scipy.stats.linregress(calibration_peak['area'], calibration_peak['concentration_mM'])
#get the slope and intercept
slope = output[0]
intercept = output[1]

print("******Calibration curve*****")
print("Slope: "+str(slope))
print("Intercept: "+str(intercept))
#save fit to file with all calibration lines
with open("data/calibration.json", "r") as jsonFile:
    data = json.load(jsonFile)
data[name+'_'+date+'_slope'] = slope
data[name+'_'+date+'_intercept']=intercept
outjson='data/calibration.json'
with open(outjson, "w") as jsonFile:
    json.dump(data, jsonFile)
print("Calibration curve saved as: "+name+'_'+date+"in"+outjson)


####
#plot the calibration curve 
####
peak_range = np.linspace(0, 1.1*calibration_peak['area'].max(),300) # Set up a range of concentrations to plot 
fit = intercept + slope * peak_range # Compute the calibration curve for plotting
plt.plot(peak_range, fit, 'k-', label='fit')
plt.plot(calibration_peak['area'],calibration_peak['concentration_mM'],'o')
plt.ylabel('concentration [mM]')
plt.xlabel('area under curve [mV*min]')
```

# Butyrate calibration


```python

#give calibration a name
name='epsilon_butyrate'
date='220331'
#specify which peak to use for calibration curve
peak_retentiontime=[35., 37.]

#from all peaks select specific peaks - according to retention time
calibration_peak = cal_peaks[(cal_peaks['retention_time'] >= peak_retentiontime[0]) & (cal_peaks['retention_time'] <= peak_retentiontime[1])]
print(calibration_peak['sample'].values)
```

```python
# get list with concentration and peak intensity
concs = []
# Loop through each sample name 
for s in calibration_peak['sample'].values:
    # Split the file name by underscores
    split = s.split('_')
    #print(split)
    ## Take only the last item that was split by an underscore and remove the mM
    concentration = split[-2].split('mM')[0]
    ## Add the concentration to the storage list as a float
    print("detected concentration: "+concentration)
    concs.append(float(concentration))
  
# Add the concentration as a new column in the data. 
calibration_peak.loc[:,'concentration_mM'] = concs #warnin
display(calibration_peak)
```

```python
#Do a fit and store the obtained calibration curve

#select what to use
#calibration_peak=calibration_peak.loc[calibration_peak['concentration_mM']!=2]

# Perform the linear regression
output = scipy.stats.linregress(calibration_peak['area'], calibration_peak['concentration_mM'])
# Get the slope and intercept
slope = output[0]
intercept = output[1]

print("******Calibration curve*****")
print("Slope: "+str(slope))
print("Intercept: "+str(intercept))
#save fit to file
with open("data/calibration.json", "r") as jsonFile:
    data = json.load(jsonFile)
data[name+'_'+date+'_slope'] = slope
data[name+'_'+date+'_intercept']=intercept

outjson='data/calibration.json'
with open(outjson, "w") as jsonFile:
    json.dump(data, jsonFile)
print("Calibration curve saved as: "+name+'_'+date+"in"+outjson)
    
####
#plot the calibration curve 
####
peak_range = np.linspace(0, 1.1*calibration_peak['area'].max(),300) # Set up a range of concentrations to plot 
fit = intercept + slope * peak_range # Compute the calibration curve for plotting

# Plot and label the data 
plt.plot(peak_range, fit, 'k-', label='fit')
plt.plot(calibration_peak['area'],calibration_peak['concentration_mM'],'o')
plt.ylabel('concentration [mM]')
plt.xlabel('intensity [mV]')
```
