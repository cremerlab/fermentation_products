# Analyze HPLC data and growth curves FECAL samples
- Go through different samples and identify peaks of specific substrates and the corresponding concentrations
- Setup to consider one growth curve at a time

```python
#!pip install cremerlab-utils
import sys, os
#sys.path.insert(0, os.path.expanduser('~/.local/lib/python3.8/site-packages'))
import glob 
import cremerlab.hplc

#load required pacakges (as always)
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.integrate as spi
from scipy.integrate import odeint #this is the module to solve ODEs
%matplotlib inline
# Load the scipy module
import scipy.stats 
import json
```

```python
#prepare everything for analysis
#set variables for analysis of HPLC data
peaks=['glucose','acetate','propionate','ethanol','lactate','succinate','butyrate','formate','lacother1','lacother2']
peaks_short=['glu','ace','prp','eth','lac','suc','but','for','lacO1','lacO2']

peak_retentiontimes=[[15.2,16],[25,26],[29,30],[32,33],[21.3,21.9],[20.6,21.2],[35,36.5],[23.,23.6],[19.8,20.3],[18.7,19.3]]
chromogram_characteristics=[peaks_short,peak_retentiontimes]

#all calibration curves have been stored in a single json file. Provide here which calibration curve to use for the different peaks
calibration=['epsilon_glucose_220331','epsilon_acetate_220321','N-C-_propionate_220610','N-C-_ethanol_220610','epsilon_lactate_220825','epsilon_succinate_220321','epsilon_butyrate_220331','N-C-_formate_220331','epsilon_lactate_220825','epsilon_lactate_220825']
with open('data/calibration.json') as file:
    cal_file = json.load(file)
cal_file

def analyze_chromatograms(listsamplenames,foldername,plot_output_chromatogram='chromatogram.pdf',plot_output_chromatogramanalysis='analysis.pdf'):
    #listsamplenames=['B.adolescentis_1']
    print("listsamplename")
    print(listsamplenames)
    
    # check chromatograms, autodetect peak positions, and add more peaks if needed
    files = glob.glob('data/'+foldername+'/converted/*.csv')
    files2 = []
    print("foldername")
    print(foldername)
    print("files in foldername")
    print(files)
    print("listsamplenames")
    print(listsamplenames)
    for file in files:
        if file.split('/')[-1][:-17] in listsamplenames:
            files2.append(file)
        else:
            pass
        
        
    print("analysis with files: ")
    print(files2)
    files=files2
    
    #plot chromatograms and identify peaks
    backgroundsubstraction=True
    backgroundsubstraction_iterations=80
    chroms, plot, peakpositions = cremerlab.hplc.batch_plot(files,time_window=[14, 37], upper_limit=50000,lower_limit=-500, detection_peakpositions='peakautodetection',backgroundsubstraction=backgroundsubstraction,backgroundsubstraction_iterations=backgroundsubstraction_iterations)
    print("autopeakposition")
    print(peakpositions)
    
    #add manualvalues to list if no closed by peak is present
    manuallist=[17,23,23.8, 25.8, 27, 31.2,33.2]
    manuallist=manuallist+[14.35833, 15.791670000000002, 17.941670000000006, 19.13333, 20.35833, 21.741670000000006, 25.13333, 29.30833, 32.44167,14.35833, 15.791670000000002, 17.941670000000006, 19.125, 20.35833, 21.65833, 23.26667, 25.125, 29.30833, 32.44167,32.43333,36.2]
    manuallist.sort()
    def add_missing_peaks(peakpositions,manuallist,mindistance=0.3):
        #go through every sample and add values if mixing
        for il in range(len(peakpositions)):
            curl=np.asarray(peakpositions[il])
            for ml in manuallist:
                idx = (np.abs(curl - ml)).argmin()
                diff=np.abs(curl[idx]-ml)
                if diff>mindistance:
                    peakpositions[il].append(ml)  
        return peakpositions

    peakpositions=add_missing_peaks(peakpositions,manuallist)

    print("Chromatograms and all identified peaks (after adding peaks manually):")
    print(peakpositions)
    #plot peak positions again, including added peaks
    chroms, plot, _ = cremerlab.hplc.batch_plot(files,time_window=[14, 37], upper_limit=5000,lower_limit=-500, plot_verticallines=peakpositions,backgroundsubstraction=backgroundsubstraction,backgroundsubstraction_iterations=backgroundsubstraction_iterations,plot_characteristictimes=chromogram_characteristics,plot_output=plot_output_chromatogram)


    #quantify intensity bellongng to peaks by fitting skewed Gaussians
    #boundpars=[0.5,0.01,0.0,-0.1,1.6,0.01,2,0.1] #set boundaries of fitting
    boundpars=[0.4,0.01,0.0,-0.1,1.5,0.01,.5,0.1]
    boundpars=[0.1,0.1,0.0,-0.1,10,0.1,5,0.1]
    #1. min. amplitude of peak height (multiplication of amplitude of detected peak position)
    #2. min. position of peak (offset of detected peak position to the left)
    #3. min. width of fitted peak
    #4. min. value of skew parameter
    #5. max. amplitude of peak height (multiplication of amplitude of detected peak position)
    #6. max. position of peak (offset of detected peak position to the right)
    #7. max. width of fitted peak
    #8. min. value of skew parameter

    #set options background correction
    backgroundsubstraction=True
    backgroundsubstraction_iterations=80
    data_chroms, data_peaks, data_plot = cremerlab.hplc.batch_process(files, time_window=[14, 42], show_viz=True,upper_limit=5000,lower_limit=-500,manual_peak_positions=peakpositions,boundpars=boundpars,buffer=100,backgroundsubstraction=backgroundsubstraction,backgroundsubstraction_iterations=backgroundsubstraction_iterations,plot_characteristictimes=chromogram_characteristics,plot_output=plot_output_chromatogramanalysis)
    print(data_peaks)
    #go through samples and find specific peaks for different substrates and store everything into a json file

    #open calibration file
    with open("data/calibration.json", "r") as jsonFile:
            cal_data = json.load(jsonFile)

    #go through each sample.....
    samplelist=data_peaks['sample'].unique()
    for sample in samplelist:
        dictout={}    
        data_selection=data_peaks.loc[data_peaks['sample']==sample]


        #check for  peaks we want to look at in that particular sample (e.g. peaks=["acetate",glucose])
        #statements below are relating the calibration data with the HPLC analysis
        for il in range(0,len(peaks)):
            peak=peaks[il]
            cal_name=calibration[il]
            slope=cal_data[cal_name+"_slope"]
            intercept=cal_data[cal_name+"_intercept"]
            peak_retentiontime=peak_retentiontimes[il]

            data_selection_currentpeak=data_selection[(data_selection['retention_time'] >= peak_retentiontime[0]) & (data_selection['retention_time'] <= peak_retentiontime[1])]
            #display(data_selection_currentpeak)
            if data_selection_currentpeak.shape[0]==0:
                print("Peak "+peak+" not found in sample "+sample)
            elif data_selection_currentpeak.shape[0]>0:
                dictout[peak]=data_selection_currentpeak['area'].values[0]*slope+intercept
                dictout[peak+"_cal"]=cal_name
                dictout['sample']=sample

            elif data_selection_currentpeak.shape[0]>1:
                morethanonepeak #check why you get more than one peak within the retention time range

        
        #display(dictout)
        with open('hplc_analysis/'+sample+'.json', 'w') as fp:
            json.dump(dictout, fp)
        # Isolate the peak used for calibration
    print(data_chroms)
    return(data_chroms)
```
