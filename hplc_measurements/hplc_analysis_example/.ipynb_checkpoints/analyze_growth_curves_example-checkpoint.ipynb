# Calibration run

To get calibration curves collect chromatograms for different samples covering a range of substrate concentrations. If the retention times for different substrates are known, different substrates can be provided in the same samples. 



```python

#Associate peaks to substrates
substratelist=["glucose","propionate","succinate","lactate","butyrate","acetate","formate","maltose"]
timewindows=[[14.8,15.5],[27.5,28],[19.5,20.2], [20.2,20.8],[33.5,35],[23.7,24.2],[21.8,22.5],[12.5,13]]
date=2308

#Folder in which samples for calibration are stored
runc="data_hplcrawdata/example_calibrationrun"

#File in which properties of samples are stored (including provided concentrations of different substrates)
filename_settings="hplc_settings_calibration_example.csv"

#Use the following options only when fine tuning parameters
analysis_window=None #[12,20] #Sets time interval to analyze. Overwrites settings from settings table if not None.
peakpositions=None #[18.7] #Sets manual peak positions. Overwrites settings from settings table if not None.
plot_time_window=None #[12,25]
cremerlab.hplc.run_calibration(runc,substratelist,timewindows,analysis_window=analysis_window,date=date,list_samplenames=None, name_prefix="YCA_Zurich",filename_settings=filename_settings,output_json='calibration.json',save_full_calibrationcurve='data_calibration',plot_time_window=plot_time_window, plot_upper_limit=40,plot_lower_limit=-2)
```

# Analyze HPLC data and growth curves
- Go through different HPLC samples collected at different timpointes of a growing culture. 
- Identify peaks of specific substrates and corresponding concentrations in each taken HPLC sample.
- Plot substrate concentrations versus OD and extract yield values. 

The script needs three different major types of data:
- The chromatograms from the HPLC runs located in a subfolder of data_hplcrawdata.
- A spreadsheet with growth curves and names of the HPLC samples taken at specific times, located in data_growthcurves. 
- The coordinates of the needed calibration curves (stored in json files e.g. calibration.json).

Below we show the analysis for one example. 

```python

#1. Set name of spreadsheet with growth curves. Spreadsheet needs to be saved as csv file in data_growthdata
runc="example_growthcurve"

#2. Decide which growth curve in spreadsheed to run. 
#culture_list="All" #Analyses all growth curves in 
culture_list=["C.aerofaciens","E.rectale08"] #the culture names need to name the corresponding column names in the spreadsheed with growth curves.
#['R.bromii', 'R.bromii2', 'P.distastonis3', 'B.fragilis3', 'B.ovatus3', 'B.vulgatus3', 'B.finegoldii3']

#3. Make sure options for HPLC settings are set. 
# Set name of spreadsheet which contains all settings to analyze the HPLC chromatograms
filename_settings="hplc_settings_example.csv"
#Make sure all hplc samples part of the growth curves are needed. In the spreadsheet set also, which calibration and peak assignment file to use. 

#For analysis, you can also temporarily overwrite the settings in the settings file and provide here manual peak positions and the time interval to use for analyze. Only do this temporarily and have final settings stored in the settings table.
analysis_window=None #[12,20] #[12,20] #Sets time interval to analyze. Overwrites settings from settings table if not None.
peakpositions=None #[15]#None #[18.7] #Sets manual peak positions. Overwrites settings from settings table if not None.
peak_assignment=None #Provided information on peak assignment

#4. Run analysis.
plot_upper_limit=5 #Provide upper limit of intensity when plotting. Autoscaling if None. 
plot_lower_limit=None #Provide lower limit of intensity when plotting. Autoscaling if None
plot_time_window=None #[20,25] #Provide time-interval to use in plots. Auto adjustment if None. 
display_fullpeaklist=True #Set True if output should provide all detected peaks. 

cremerlab.hplc.process_growth_curve(runc,folder_hplcdata=None,analysis_window=analysis_window,peakpositions=peakpositions,culture_list=culture_list,filename_settings=filename_settings,folder_output="output_growthanalysis", plot_upper_limit=plot_upper_limit,plot_lower_limit=plot_lower_limit,display_fullpeaklist=display_fullpeaklist,plot_time_window=plot_time_window)
```
