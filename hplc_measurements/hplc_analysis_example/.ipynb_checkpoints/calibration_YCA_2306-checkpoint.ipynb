# Acetate calibration

```python
#give calibration a name
name='epsilon_acetate'
date='220321'
#specify which peak to use for calibration curve
peak_retentiontime=[25., 25.5]

#from all peaks select specific peaks - according to retention time
calibration_peak = cal_peaks[(cal_peaks['retention_time'] >= peak_retentiontime[0]) & (cal_peaks['retention_time'] <= peak_retentiontime[1])]
print(calibration_peak['sample'].values)
```

```python
# Set up a list to store the concentrations
concs = []
# Loop through each sample name to get information about used concentration
# file name e.g. scfa_2mM001_chromatogram.csv
for s in calibration_peak['sample'].values:
    # Split the file name by underscores
    split = s.split('_') #output: ['scfa','2mM001','chromatogram.csv']
    #print(split)
    ## Take only the last item that was split by an underscore and remove the mM
    concentration = split[-2].split('mM')[0]
    ## Add the concentration to the storage list as a float
    print("detected concentration: "+concentration)
    concs.append(float(concentration))
# Add the concentration as a new column in the data. 
calibration_peak.loc[:,'concentration_mM'] = concs #warnin
display(calibration_peak)
```

```python
#Do a fit and store the obtained calibration curve

#select what to use
#calibration_peak=calibration_peak.loc[calibration_peak['concentration_mM']>0.5]

# Perform the linear regression
output = scipy.stats.linregress(calibration_peak['area'], calibration_peak['concentration_mM'])
# Get the slope and intercept
slope = output[0]
intercept = output[1]

print("******Calibration curve*****")
print("Slope: "+str(slope))
print("Intercept: "+str(intercept))
#save fit to file
with open("data/calibration.json", "r") as jsonFile:
    data = json.load(jsonFile)
data[name+'_'+date+'_slope'] = slope
data[name+'_'+date+'_intercept']=intercept

outjson='data/calibration.json'
with open(outjson, "w") as jsonFile:
    json.dump(data, jsonFile)
print("Calibration curve saved as: "+outjson)
    
####
#plot the calibration curve 
####
peak_range = np.linspace(0, 1.1*calibration_peak['area'].max(),300) # Set up a range of concentrations to plot 
fit = intercept + slope * peak_range # Compute the calibration curve for plotting

# Plot and label the data 
plt.plot(peak_range, fit, 'k-', label='fit')
plt.plot(calibration_peak['area'],calibration_peak['concentration_mM'],'o')
plt.ylabel('concentration [mM]')
plt.xlabel('AUC of fitted peak [mV min]')
plt.savefig('plot_output/cal_'+name+'_'+date+'.pdf')
```
